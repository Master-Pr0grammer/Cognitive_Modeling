```python
import numpy as np
import matplotlib.pyplot as plt
```

P(D = True) = 0.01 -> Probability of having disease

P(T=True | D=True) = 0.95 -> (Sensitivity) Probability of testing positive, given you have the disease

P(T=False | D=False) = 0.9 -> (Specificity) Probability of testing negative, given you dont have the disease



Posterior = P(D=True | T=True)

```python
# Part 1
# Y = Posterior
# X = prior P(D)
#	sensitivity, specificity are const

# A = D
# B = T

y_data_1 = np.zeros(100)
for i in range(1,101):
	P_D = (i/100)
	P_T = 0.95*P_D + (1-0.9)*(1-P_D)
	y_data_1[i-1] = P_D * (0.95 / P_T)

plt.plot(y_data_1)
plt.title('Posterior Vs Prior')
plt.ylabel('Posterior: P(D=True | T=True)')
plt.xlabel('Prior: P(D=True)')
plt.show()
```

```python
# Part 2
# Y = Posterior
# X = test sensitivity P(T=True | D=True)
#	prior, specificity are const

# A = D
# B = T

P_T = 0.01 * 0.95
y_data_2 = np.zeros(100)
for i in range(1,101):
	y_data_2[i-1] = 0.01 * ((i/100) / P_T)

plt.plot(y_data_2)
plt.title('Posterior Vs Test Sensitivity')
plt.ylabel('Posterior: P(D=True | T=True)')
plt.xlabel('Test Sensitivity: P(T=True | D=True)')
plt.show()
```

```python
# Part 3
# Y = Posterior
# X = test specificity P(T=True | D=True)
#	sensitivity, specificity are const

# A = D
# B = T

y_data_3 = np.zeros(100)
for i in range(1,101):
	P_T = 0.95*0.01 + (1-(i/100))*(1-0.01)
	y_data_3[i-1] = 0.01 * (0.95 / P_T)

plt.plot(y_data_3)
plt.title('Posterior Vs Test specificity')
plt.ylabel('Posterior: P(D=True | T=True)')
plt.xlabel('Test specificity: P(T=False | D=False)')
plt.show()
```
