```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

# Number of points to generate
num_pts = 10000

# Generate points
pts = np.random.random(size=(num_pts,2))
inside_circle = (1 >= (pts[:,0]**2 + pts[:,1]**2)).astype(np.int16)
```

```python
# Get the data to plot
pi_estimates = []
rolling_sum = 0
for i in range(num_pts):
	rolling_sum+=inside_circle[i]
	pi_estimates.append(4*rolling_sum/(i+1))

error = np.abs(np.array(pi_estimates) - np.pi)

# Plot data
plt.plot(error)
plt.title('Error Vs. Number of Points')
plt.xlabel('Number of Points')
plt.ylabel('Error')
```
