```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given hyperparameters for the prior
mu_0 = 0
sigma_0_squared = 1

# Known variance for the likelihood and generating data
sigma_squared = 1
true_mu = 1
num_samples = 50
data_points = np.random.normal(true_mu, np.sqrt(sigma_squared), num_samples)

# Calculating the posterior parameters
n = len(data_points)
mu_n = (mu_0 / sigma_0_squared + data_points.sum() / sigma_squared) / (1 / sigma_0_squared + n / sigma_squared)
sigma_n_squared = 1 / (1 / sigma_0_squared + n / sigma_squared)

# Simulating a large number of points from the prior and posterior for histogram representation
simulated_prior = np.random.normal(mu_0, np.sqrt(sigma_0_squared), 1000)
simulated_posterior = np.random.normal(mu_n, np.sqrt(sigma_n_squared), 1000)

# Plotting the histograms
plt.figure(figsize=(14, 7))

# Histogram for the prior distribution
plt.hist(simulated_prior, bins=30, alpha=0.5, density=True, label='Simulated Prior Distribution')

# Histogram for the posterior distribution
plt.hist(simulated_posterior, bins=30, alpha=0.5, density=True, label='Simulated Posterior Distribution')

# Histogram for the observed data points
plt.hist(data_points, bins=num_samples, density=True, alpha=0.6, color='gray', label='Observed Data Points')

plt.title('Histograms of Simulated Prior and Posterior Distributions with Observed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
