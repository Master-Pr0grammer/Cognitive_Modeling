```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# 1D Case
x = np.random.normal(loc=0, scale=1, size=10000)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
sb.histplot(x, ax=ax)
```

```python
# 2d Case
normal_curve_center = [0, 0]
covariance = [[1, 2], [2, 1]]
x = np.random.multivariate_normal(normal_curve_center, covariance, size=10000)

# Graph
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(x[:,0], x[:,1])
_ = sb.kdeplot(x=x[:,0], y=x[:,1], fill=True)
```

```python
normal_curve_center = [0, 0]
covariance = [[1, 2], [2, 1]]
scale = 1
mvn = stats.multivariate_normal(normal_curve_center, scale)

x = mvn.rvs(size=100)
pdfs = mvn.pdf(x)

pdfs.shape
```
